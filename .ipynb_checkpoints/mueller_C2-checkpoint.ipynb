```python
import os
import numpy as np
import matplotlib
from  matplotlib import pyplot as plt
import librosa
from ipywidgets import interact, fixed, FloatSlider
import IPython.display as ipd
%matplotlib inline

Fs = 128
duration = 10
omega1 = 1
omega2 = 5
N = int(duration * Fs)
t = np.arange(N) / Fs
t1 = t[:N//2]
t2 = t[N//2:]

x1 = 1.0 * np.sin(2 * np.pi * omega1 * t)
x2 = 0.7 * np.sin(2 * np.pi * omega2 * t)
x = x1 + x2
# x = np.concatenate((x1, x2))

fig = plt.figure(figsize=(8, 2))
ax = fig.add_subplot(1, 2, 1)
plt.plot(t, x, c='k')
plt.xlim([min(t), max(t)])
plt.xlabel('Time (seconds)', c='white')
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

ax1 = fig.add_subplot(1, 2, 2)
X = np.abs(np.fft.fft(x)) / Fs
freq = np.fft.fftfreq(N, d=1/Fs)
X = X[:N//2]
freq = freq[:N//2]
plt.plot(freq, X, c='k')
plt.xlim([0, 7])
plt.ylim([0, 7])
plt.xlabel('Frequency (Hz)', c='white')
ax1.tick_params(axis='x', colors='white')
ax1.tick_params(axis='y', colors='white')

plt.tight_layout()
```


\begin{align}
T&= 10s \\
x &= \begin{cases}
x_1 = \sin(2\pi t) & 0\leq t<5 \\
x_2 = 0.7\sin(10\pi t) & 5\leq t<10
\end{cases}
\end{align}
